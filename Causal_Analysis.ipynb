```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import gc # Garbage Collector interface
import sys # To exit script gracefully

# --- 0. Configuration ---

# --- File Path ---
# IMPORTANT: Replace with the correct path to your data file
file_path = 'C:/Users/pdudas/Desktop/Chatbot Stuff/HealthcodeErrors_202504161531.csv'

# --- EDA Settings ---
N_TOP_HEALTHCODES = 15      # Number of top Healthcodes to focus on in bivariate plots
M_TOP_TYPES = 15            # Number of top Types to focus on
K_TOP_TRACK_SECTIONS = 15 # Number of top Track Sections to focus on

# --- Column Names (Update if necessary based on your actual file) ---
# IMPORTANT: Verify this matches your column name EXACTLY
# Check for spaces, capitalization mismatches in your CSV header.
track_section_col = 'TrackSection' # <--- *** POTENTIAL FIX AREA: VERIFY THIS NAME ***
timestamp_col = 'Timestamp'
ingestion_date_col = 'IngestionDate'
message_col = 'Message'
healthcode_col = 'Healthcode'
type_col = 'Type'
vehicle_col = 'VobcId' # Adjusted based on user's last data snippet
# Add other key columns if needed for specific type conversions later
numerical_cols_base = ['VccNum', 'LoopNum', 'SegmentId', 'XOverNum'] # Adjusted based on user's last data snippet

# --- Date/Time Parsing Formats (ADJUSTED based on sample data) ---
# Both timestamp and ingestionDate follow the format: DD/MM/YYYY  HH:MM:SS
# Note the two spaces between the date and the time.
datetime_format = '%d/%m/%Y  %H:%M:%S'

# --- Helper Function to Reduce Memory Usage of DataFrame ---
def reduce_mem_usage(df, verbose=True):
    """ Iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.
    """
    start_mem = df.memory_usage(deep=True).sum() / 1024**2
    if verbose:
        print(f'Memory usage of dataframe is {start_mem:.2f} MB')

    for col in df.columns:
        col_type = df[col].dtype

        # Skip object, category, and datetime columns
        if col_type == object or col_type.name == 'category' or 'datetime' in str(col_type):
            continue

        # Handle numeric types
        c_min = df[col].min()
        c_max = df[col].max()
        if str(col_type)[:3] == 'int':
            if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                df[col] = df[col].astype(np.int8)
            elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                df[col] = df[col].astype(np.int16)
            elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                df[col] = df[col].astype(np.int32)
            elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                df[col] = df[col].astype(np.int64)
        else: # float
            # Check if float can be represented as int without loss
            # Be careful with this check on large floats or where precision matters
            is_int_repr = False
            if df[col].isnull().sum() == 0: # Cannot compare NaNs
                 if (df[col] == df[col].astype(np.int64)).all():
                        is_int_repr = True

            if is_int_repr:
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            # Check float precision
            elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                df[col] = df[col].astype(np.float32)
            # Keep float64 if requires high precision or range

    end_mem = df.memory_usage(deep=True).sum() / 1024**2
    if verbose:
        print(f'Memory usage after optimization: {end_mem:.2f} MB')
        if start_mem > 0:
             print(f'Decreased by {(start_mem - end_mem) / start_mem * 100:.1f}%')

    return df

# --- 1. Load Data ---
print(f"\n--- 1. Loading Data from {file_path} ---")
try:
    df = pd.read_csv(file_path)
    print(f"Successfully loaded data. Shape: {df.shape}")
    print(f"Initial columns found: {list(df.columns)}")
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    sys.exit()
except Exception as e:
    print(f"Error loading data: {e}")
    sys.exit()

# --- Check that the defined track_section_col actually exists ---
if track_section_col not in df.columns:
     print(f"\nCRITICAL ERROR: The specified track section column '{track_section_col}' was not found in the loaded data.")
     print(f"Please check the 'track_section_col' variable definition near the top of the script.")
     sys.exit()

# Apply memory reduction early
df = reduce_mem_usage(df, verbose=True)

# --- 2. Initial Inspection ---
print("\n--- 2. Initial Inspection ---")
print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Info (Columns, Non-Null Counts, Data Types, Memory):")
df.info(memory_usage='deep')

print("\nBasic Descriptive Statistics:")
# Describe might fail on category types if not handled correctly by pandas version
try:
    print(df.describe(include='all'))
except TypeError:
     print(df.describe(include=[np.number])) # Describe only numeric if 'all' fails
     print(df.describe(include=['object', 'category'])) # Describe non-numeric separately


# --- Check required columns needed for later steps exist ---
# Adjusted columns based on latest snippet & definition block
required_cols_list = [track_section_col, timestamp_col, ingestion_date_col, message_col, healthcode_col, type_col, vehicle_col] + numerical_cols_base
missing_req_cols = [col for col in required_cols_list if col not in df.columns]
if missing_req_cols:
    print(f"\nError: Some expected columns missing from DataFrame after loading: {missing_req_cols}")
    print("Processing stopped. Please check input file and column name definitions.")
    sys.exit()


# --- 3. Basic Data Cleaning / Preprocessing ---
print("\n--- 3. Data Cleaning / Preprocessing ---")

# Drop Redundant Message Column (as confirmed by user)
if message_col in df.columns:
    print(f"\nDropping redundant column: {message_col}")
    try:
        df.drop(columns=[message_col], inplace=True)
        gc.collect() # Explicitly call garbage collector
    except KeyError:
        print(f"Column {message_col} already removed or does not exist.")


# Handle Missing Values
print("\nMissing values per column (before handling):")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
# --- Strategy: Drop rows with NA in critical columns ---
# Define critical columns needed for core analysis
critical_na_cols = [healthcode_col, type_col, track_section_col, timestamp_col, ingestion_date_col, vehicle_col]
# Ensure these critical columns actually exist before trying to drop NA based on them
critical_na_cols = [col for col in critical_na_cols if col in df.columns]
if critical_na_cols:
    print(f"\nDropping rows with NAs in any of: {critical_na_cols}...")
    start_rows = df.shape[0]
    df.dropna(subset=critical_na_cols, inplace=True)
    print(f"Dropped {start_rows - df.shape[0]} rows.")
    if df.empty:
        print("DataFrame is empty after dropping NA rows. Stopping.")
        sys.exit()
else:
    print("Skipping NA drop, no critical columns specified or found.")


# Handle Duplicate Rows
duplicate_rows = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")
if duplicate_rows > 0:
    print("Dropping duplicate rows...")
    df.drop_duplicates(inplace=True)
    print(f"DataFrame shape after dropping duplicates: {df.shape}")
    gc.collect()

# Convert Data Types (Date/Time)
print("\nConverting date/time columns...")
datetime_conversion_successful = False
# Check if the original columns exist before attempting conversion
if timestamp_col in df.columns and ingestion_date_col in df.columns:
    try:
        print(f"Attempting to parse '{timestamp_col}' into 'EventDateTime' using format: '{datetime_format}'")
        # Parse the Timestamp column directly using the specified format
        df['EventDateTime'] = pd.to_datetime(df[timestamp_col], format=datetime_format, errors='coerce')

        datetime_nulls = df['EventDateTime'].isnull().sum()
        if datetime_nulls > 0:
            print(f"Warning: {datetime_nulls} rows resulted in NaT (Not a Time) during '{timestamp_col}' parsing.")
            print(f"This could be due to values in '{timestamp_col}' not matching the format '{datetime_format}'.")
            # Consider dropping or investigating these rows if they are problematic for analysis
            # original_rows = df.shape[0]
            # df.dropna(subset=['EventDateTime'], inplace=True)
            # print(f"Dropped {original_rows - df.shape[0]} rows with NaT EventDateTime.")
            # if df.empty:
            #    print("DataFrame is empty after dropping NaT EventDateTime rows. Stopping.")
            #    sys.exit()

        # Check if conversion was mostly successful (e.g., less than 50% NaTs or adjust threshold)
        if df['EventDateTime'].notnull().sum() > 0: # Check if at least some values were parsed
             print(f"Successfully created 'EventDateTime' from '{timestamp_col}'.")
             # Drop the original timestamp and ingestion date columns now
             print(f"Dropping original columns: '{timestamp_col}' and '{ingestion_date_col}'")
             df.drop(columns=[timestamp_col, ingestion_date_col], inplace=True)
             datetime_conversion_successful = True
        else:
             print(f"ERROR: All values failed to parse for '{timestamp_col}'. 'EventDateTime' column is all NaT.")
             # Drop the potentially created all-NaT column
             if 'EventDateTime' in df.columns:
                 df.drop(columns=['EventDateTime'], inplace=True)
             print("Skipping further datetime dependent steps.")


    except ValueError as ve:
         print(f"CRITICAL ERROR during date/time conversion: {ve}")
         print(f"This often means the data in '{timestamp_col}' does NOT match the expected format '{datetime_format}'.")
         print("Please verify the format string and the actual data in the CSV file.")
         print("Skipping datetime creation and dependent steps.")
    except Exception as e:
        print(f"Unexpected Error creating 'EventDateTime': {e}")
        print("Skipping datetime creation and dependent steps.")
    finally:
        gc.collect()
else:
     print(f"Skipping datetime conversion because one or both source columns ('{timestamp_col}', '{ingestion_date_col}') were not found in the DataFrame.")


# Convert Categorical & Optimize Numerics
print("\nConverting types for memory efficiency (Categorical)...")
# --- FIX AREA: Ensure Track section is handled before converting to category ---
categorical_to_convert = [healthcode_col, type_col, track_section_col, vehicle_col]
for col in categorical_to_convert:
    if col in df.columns:
        try:
            # If column is numeric (like Track Section might be), convert to string first
            if pd.api.types.is_numeric_dtype(df[col]):
                 print(f"Converting numeric column '{col}' to string before categorizing.")
                 df[col] = df[col].astype(str)

            # Now convert to category
            print(f"Converting column '{col}' to category type.")
            df[col] = df[col].astype('category')
        except Exception as e:
            print(f"Error converting column '{col}' to category: {e}")
            print(f"Skipping category conversion for {col}.")

print("\nConverting types for memory efficiency (Numeric)...")
# Downcast remaining numerical columns
df = reduce_mem_usage(df, verbose=False) # Run again to downcast numerics

# --- 4. Memory Check After Cleaning ---
print("\n--- 4. Memory Usage After Cleaning ---")
df.info(memory_usage='deep')
gc.collect()

# --- 5. Univariate Analysis ---
print("\n--- 5. Univariate Analysis ---")

# Update numerical_cols list based on actual columns remaining and their types
numerical_cols = [col for col in numerical_cols_base if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]

# Plotting distributions for numerical columns
print("\nPlotting distributions for numerical columns...")
for col in numerical_cols:
    if df[col].isnull().all(): # Skip if column is all NaN after cleaning
        print(f"Skipping plot for {col} as it contains only NaN values.")
        continue
    try:
        plt.figure(figsize=(10, 4))
        plt.suptitle(f'Univariate Analysis: {col}', fontsize=14)

        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True, bins=30) # Adjust bins as needed
        plt.title(f'Histogram')

        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout
        plt.show()
    except Exception as e:
        print(f"Could not generate plots for {col}: {e}")
    finally:
        plt.clf(); plt.close(); gc.collect()


# Plotting value counts for categorical columns
print("\nPlotting value counts for categorical columns...")
# Use the list defined during type conversion
categorical_cols_present = [col for col in categorical_to_convert if col in df.columns and df[col].dtype.name == 'category']

for col in categorical_cols_present:
    try:
        # Determine Top N based on settings or number of unique values
        num_unique = df[col].nunique()
        if col == healthcode_col: top_n = min(N_TOP_HEALTHCODES * 2, num_unique, 50) # Limit max display
        elif col == type_col: top_n = min(M_TOP_TYPES * 2, num_unique, 50)
        elif col == track_section_col: top_n = min(K_TOP_TRACK_SECTIONS * 2, num_unique, 50)
        else: top_n = min(20, num_unique, 50) # Limit for others like VehicleId

        print(f"\nAnalyzing top {top_n} categories for {col}...")
        if num_unique == 0:
             print(f"Skipping {col} - no unique values found (column might be empty).")
             continue

        value_counts = df[col].value_counts()
        top_categories = value_counts.nlargest(top_n)

        if top_categories.empty:
            print(f"Skipping {col} - No data to plot after nlargest.")
            continue

        plt.figure(figsize=(max(12, top_n * 0.4), 7)) # Adjust width, increase height
        sns.barplot(x=top_categories.index.astype(str), y=top_categories.values, order=top_categories.index.astype(str))
        plt.title(f'Top {top_n} Value Counts for {col} (Total Unique: {num_unique})')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.xticks(rotation=90, ha='center') # Use 90 deg rotation for potentially many cats
        plt.tight_layout()
        plt.show()
        print(f"Value counts (Top {top_n}):")
        print(top_categories)
    except Exception as e:
        print(f"Could not generate value count plot for {col}: {e}")
    finally:
        plt.clf(); plt.close(); gc.collect()


# Analyze Datetime Column
# **** MODIFICATION: Check if datetime_conversion_successful flag is True ****
if datetime_conversion_successful and 'EventDateTime' in df.columns and df['EventDateTime'].notna().any():
    print("\nAnalyzing EventDateTime column...")
    try:
        plt.figure(figsize=(15, 5))
        # Resample requires a DatetimeIndex
        event_counts_daily = df.set_index('EventDateTime')['EventDateTime'].resample('D').size()
        event_counts_daily.plot()
        plt.title('Number of Events per Day')
        plt.ylabel('Number of Events')
        plt.xlabel('Date')
        plt.tight_layout()
        plt.show()
        del event_counts_daily # Clean up resampled series
    except Exception as e:
        print(f"Could not plot time series data: {e}")
    finally:
        plt.clf(); plt.close(); gc.collect()
elif 'EventDateTime' not in df.columns:
     print("\nSkipping EventDateTime analysis because 'EventDateTime' column was not created (likely due to parsing issues).")
else:
     print("\nSkipping EventDateTime analysis because the column exists but contains only NaT values.")


# --- 6. Identify Top Categories for Bivariate ---
print("\n--- 6. Identifying Top Categories for Bivariate Analysis ---")

top_healthcodes = None
top_types = None
top_track_sections = None
track_section_analysis_possible = False # Flag to control plotting

try:
    if healthcode_col in df.columns and df[healthcode_col].dtype.name == 'category':
        top_healthcodes = df[healthcode_col].value_counts().nlargest(N_TOP_HEALTHCODES).index
        print(f"- Top {N_TOP_HEALTHCODES} Healthcodes identified: {list(top_healthcodes)}")
    else:
        print(f"Warning: Cannot identify top Healthcodes. Column '{healthcode_col}' missing or not categorical.")

    if type_col in df.columns and df[type_col].dtype.name == 'category':
        top_types = df[type_col].value_counts().nlargest(M_TOP_TYPES).index
        print(f"- Top {M_TOP_TYPES} Types identified: {list(top_types)}")
    else:
        print(f"Warning: Cannot identify top Types. Column '{type_col}' missing or not categorical.")

    # --- FIX AREA: Specific check for Track section ---
    if track_section_col in df.columns and df[track_section_col].dtype.name == 'category':
        top_track_sections = df[track_section_col].value_counts().nlargest(K_TOP_TRACK_SECTIONS).index
        print(f"- Top {K_TOP_TRACK_SECTIONS} Track sections identified: {list(top_track_sections)}")
        track_section_analysis_possible = True # Set flag to True
    else:
        print(f"Warning: Cannot identify top Track sections. Column '{track_section_col}' missing or not categorical.")
        print("Plots involving Track Section will be skipped.")
        track_section_analysis_possible = False # Keep flag False

except Exception as e:
    print(f"Error identifying top categories for bivariate analysis: {e}")
    print("Bivariate analysis may be incomplete.")
    # Decide if script should exit here depending on severity
    # sys.exit()

# Check if we have the necessary top categories identified
# bivariate_possible = top_healthcodes is not None and top_types is not None and track_section_analysis_possible # This wasn't used later, can be removed if desired

# --- 7. Bivariate Analysis ---
print("\n--- 7. Bivariate Analysis ---")

# Numerical vs. Numerical: Correlation Matrix
if len(numerical_cols) > 1:
    print("\nCalculating correlation matrix for numerical columns...")
    try:
        correlation_matrix = df[numerical_cols].corr()
        plt.figure(figsize=(8, 6))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Correlation Matrix of Numerical Variables')
        plt.show()
        print(correlation_matrix)
        del correlation_matrix # Clean up
    except Exception as e:
        print(f"Could not calculate or plot correlation matrix: {e}")
    finally:
        plt.clf(); plt.close(); gc.collect()

# --- Categorical vs. Numerical Example: Boxplots ---
if 'LoopNum' in numerical_cols and top_healthcodes is not None:
    print(f"\nAnalyzing LoopNum distribution per Top {N_TOP_HEALTHCODES} Healthcodes...")
    try:
        # Filter *only* for plotting
        df_filtered_plot = df[df[healthcode_col].isin(top_healthcodes)].copy()
        # Ensure the order for plotting using Categorical
        df_filtered_plot[healthcode_col] = pd.Categorical(df_filtered_plot[healthcode_col], categories=top_healthcodes, ordered=True)

        plt.figure(figsize=(max(12, N_TOP_HEALTHCODES * 0.8), 7))
        sns.boxplot(data=df_filtered_plot, x=healthcode_col, y='LoopNum') # Order is set by categorical type
        plt.title(f'LoopNum Distribution by Top {N_TOP_HEALTHCODES} Healthcodes')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        del df_filtered_plot # Clean up filtered df
    except Exception as e:
        print(f"Could not generate boxplot for LoopNum vs Healthcode: {e}")
    finally:
        plt.clf(); plt.close(); gc.collect()

# --- Categorical vs. Categorical ---

# Updated Function to generate and plot crosstabs
def plot_crosstab(df_data, index_col, columns_col, top_index_cats, top_column_cats, N_index, M_columns):
    # Check if input categories are valid
    if top_index_cats is None or top_column_cats is None:
        print(f"Skipping plot for {index_col} vs {columns_col} due to missing top category lists.")
        return
    if not (index_col in df_data.columns and columns_col in df_data.columns):
        print(f"Skipping plot for {index_col} vs {columns_col} as one or both columns not found in DataFrame.")
        return
    if df_data[index_col].dtype.name != 'category' or df_data[columns_col].dtype.name != 'category':
        print(f"Skipping plot for {index_col} vs {columns_col} as one or both columns are not categorical type.")
        return

    print(f"\n--- Analyzing Relationship: {index_col} vs {columns_col} ---")

    # Filter DataFrame efficiently using .isin()
    idx_filter = df_data[index_col].isin(top_index_cats)
    col_filter = df_data[columns_col].isin(top_column_cats)
    df_filtered = df_data[idx_filter & col_filter].copy()
    
    if not df_filtered.empty:
        try:
            # Set categories for consistent ordering and include all top cats even if some combinations are zero
            df_filtered[index_col] = pd.Categorical(df_filtered[index_col], categories=top_index_cats, ordered=True)
            df_filtered[columns_col] = pd.Categorical(df_filtered[columns_col], categories=top_column_cats, ordered=True)

            # --- Crosstab Counts ---
            print(f"Calculating Crosstab (Counts)...")
            crosstab_counts = pd.crosstab(df_filtered[index_col], df_filtered[columns_col], dropna=False)
            print(crosstab_counts)
            plt.figure(figsize=(max(10, M_columns*0.7), max(8, N_index*0.5)))
            sns.heatmap(crosstab_counts, annot=True, fmt='d', cmap='viridis', linewidths=.5)
            plt.title(f'Counts: Top {N_index} {index_col} per Top {M_columns} {columns_col}')
            plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
            plt.tight_layout(); plt.show()
            plt.clf(); plt.close()

            # --- Crosstab Normalized ---
            print(f"Calculating Crosstab (% within {columns_col})...")
            crosstab_norm = pd.crosstab(df_filtered[index_col], df_filtered[columns_col], normalize='columns', dropna=False) * 100
            print(crosstab_norm.round(1))
            plt.figure(figsize=(max(10, M_columns*0.7), max(8, N_index*0.5)))
            sns.heatmap(crosstab_norm, annot=True, fmt='.1f', cmap='viridis', linewidths=.5)
            plt.title(f'% of {index_col} within each {columns_col} (Top {N_index} / Top {M_columns})')
            plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
            plt.tight_layout(); plt.show()
            plt.clf(); plt.close()

            del crosstab_counts, crosstab_norm # Clean up
        except Exception as e:
            print(f"Error generating crosstab/heatmap for {index_col} vs {columns_col}: {e}")
    else:
        print(f"No overlapping data found for the selected top {index_col} and {columns_col}.")
    del df_filtered # Clean up
    gc.collect()


# Generate the crosstab plots conditionally
if top_healthcodes is not None and top_types is not None:
    plot_crosstab(df, healthcode_col, type_col, top_healthcodes, top_types, N_TOP_HEALTHCODES, M_TOP_TYPES)
else:
    print("\nSkipping Healthcode vs Type analysis due to missing top categories.")

# --- FIX AREA: Conditional execution for Track section ---
if track_section_analysis_possible and top_healthcodes is not None and top_track_sections is not None:
    plot_crosstab(df, healthcode_col, track_section_col, top_healthcodes, top_track_sections, N_TOP_HEALTHCODES, K_TOP_TRACK_SECTIONS)
else:
    print("\nSkipping Healthcode vs Track section analysis (Track section analysis not possible or missing top categories).")

if track_section_analysis_possible and top_types is not None and top_track_sections is not None:
    plot_crosstab(df, type_col, track_section_col, top_types, top_track_sections, M_TOP_TYPES, K_TOP_TRACK_SECTIONS)
else:
    print("\nSkipping Type vs Track section analysis (Track section analysis not possible or missing top categories).")


print("\n--- EDA Script Finished ---")
```

```python
import pandas as pd
# import matplotlib.pyplot as plt # Plotting disabled
# import seaborn as sns # Plotting disabled
import numpy as np
import gc # Garbage Collector interface
import sys # To exit script gracefully
import os # To handle file paths

# --- 0. Configuration ---

# --- File Paths ---
# IMPORTANT: Replace with the correct path to your data file
file_path = 'C:/Users/pdudas/Desktop/Chatbot Stuff/HealthcodeErrors_202504161531.csv'
# IMPORTANT: Define where to save the processed data
output_dir = os.path.dirname(file_path) # Save in the same directory as input by default
output_filename = 'processed_healthcode_data.pkl' # Using pickle to preserve dtypes
output_file_path = os.path.join(output_dir, output_filename)


# --- EDA Settings ---
N_TOP_HEALTHCODES = 15      # Number of top Healthcodes to focus on
M_TOP_TYPES = 15            # Number of top Types to focus on
K_TOP_TRACK_SECTIONS = 15 # Number of top Track Sections to focus on

# --- Column Names (Update if necessary based on your actual file) ---
# IMPORTANT: Verify this matches your column name EXACTLY
track_section_col = 'TrackSection' # <--- *** POTENTIAL FIX AREA: VERIFY THIS NAME ***
timestamp_col = 'Timestamp'
ingestion_date_col = 'IngestionDate'
message_col = 'Message' # Keep Message if needed for extraction, otherwise it's dropped
healthcode_col = 'Healthcode'
type_col = 'Type'
vehicle_col = 'VobcId' # Adjusted based on user's last data snippet
numerical_cols_base = ['VccNum', 'LoopNum', 'SegmentId', 'XOverNum'] # Adjusted based on user's last data snippet

# --- Date/Time Parsing Formats (ADJUST THESE CAREFULLY!) ---
# This is your BEST GUESS for the format. The code will try this first.
# If this fails, it will try to automatically infer the format.
# Example format: Day/Month/Year<TWO SPACES>Hour:Minute:Second
# *** LOOK AT THE PRINTED RAW DATA AND UPDATE THIS IF NEEDED ***
datetime_format = '%d/%m/%Y  %H:%M:%S'


# --- Helper Function to Reduce Memory Usage of DataFrame ---
def reduce_mem_usage(df, verbose=True):
    """ Iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.
    """
    start_mem = df.memory_usage(deep=True).sum() / 1024**2
    if verbose:
        print(f'Memory usage of dataframe is {start_mem:.2f} MB')

    for col in df.columns:
        col_type = df[col].dtype

        # Skip object, category, and datetime columns
        if col_type == object or col_type.name == 'category' or 'datetime' in str(col_type):
            continue

        # Handle numeric types
        # Check for NaN before min/max if column allows nulls
        if df[col].isnull().any():
             # Skip if all values are NaN
             if df[col].notna().sum() == 0:
                 continue
             c_min = df[col].min(skipna=True)
             c_max = df[col].max(skipna=True)
        else:
            # Handle case where column might be empty after filtering
            if df[col].empty:
                continue
            c_min = df[col].min()
            c_max = df[col].max()

        # Skip if min/max could not be determined (e.g., all NaNs)
        if pd.isna(c_min) or pd.isna(c_max):
             continue

        if str(col_type)[:3] == 'int':
            if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max: # Use >= and <=
                df[col] = df[col].astype(np.int8)
            elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                df[col] = df[col].astype(np.int16)
            elif c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max:
                df[col] = df[col].astype(np.int32)
            elif c_min >= np.iinfo(np.int64).min and c_max <= np.iinfo(np.int64).max:
                df[col] = df[col].astype(np.int64)
        elif str(col_type)[:5] == 'float': # Check float types
            # Check if float can be represented as int without loss
            is_int_repr = False
            # Check only non-null values for integer representation
            non_null_series = df[col].dropna()
            if not non_null_series.empty:
                try:
                    # Use np.isclose for float comparison to handle potential precision issues
                    if np.isclose(non_null_series, non_null_series.astype(np.int64)).all():
                        is_int_repr = True
                except (OverflowError, TypeError):
                    # Handle cases where conversion to int64 might fail
                    pass

            if is_int_repr:
                # Convert to integer types based on range
                if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min >= np.iinfo(np.int64).min and c_max <= np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            # Check float precision only if not convertible to int
            elif c_min >= np.finfo(np.float32).min and c_max <= np.finfo(np.float32).max:
                df[col] = df[col].astype(np.float32)
            # Keep float64 otherwise

    end_mem = df.memory_usage(deep=True).sum() / 1024**2
    if verbose:
        print(f'Memory usage after optimization: {end_mem:.2f} MB')
        if start_mem > 0 and end_mem > 0: # Avoid division by zero if start_mem is 0
             print(f'Decreased by {(start_mem - end_mem) / start_mem * 100:.1f}%')

    return df

# --- 1. Load Data ---
print(f"\n--- 1. Loading Data from {file_path} ---")
try:
    # Load the CSV, consider specifying dtypes if known for efficiency
    df = pd.read_csv(file_path)
    print(f"Successfully loaded data. Shape: {df.shape}")
    print(f"Initial columns found: {list(df.columns)}")
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    sys.exit()
except Exception as e:
    print(f"Error loading data: {e}")
    sys.exit()

# --- Check that the defined track_section_col actually exists ---
if track_section_col not in df.columns:
     print(f"\nCRITICAL ERROR: The specified track section column '{track_section_col}' was not found in the loaded data.")
     print(f"Please check the 'track_section_col' variable definition near the top of the script.")
     sys.exit()

# Apply memory reduction early
df = reduce_mem_usage(df, verbose=True)

# --- 2. Initial Inspection ---
print("\n--- 2. Initial Inspection ---")
print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Info (Columns, Non-Null Counts, Data Types, Memory):")
df.info(memory_usage='deep')

print("\nBasic Descriptive Statistics:")
# Describe might fail on category types if not handled correctly by pandas version
try:
    print(df.describe(include='all'))
except TypeError:
     print(df.describe(include=[np.number])) # Describe only numeric if 'all' fails
     print(df.describe(include=['object', 'category'])) # Describe non-numeric separately


# --- Check required columns needed for later steps exist ---
# Adjusted columns based on latest snippet & definition block
required_cols_list = [track_section_col, timestamp_col, ingestion_date_col, message_col, healthcode_col, type_col, vehicle_col] + numerical_cols_base
missing_req_cols = [col for col in required_cols_list if col not in df.columns]
if missing_req_cols:
    print(f"\nError: Some expected columns missing from DataFrame after loading: {missing_req_cols}")
    print("Processing stopped. Please check input file and column name definitions.")
    sys.exit()


# --- 3. Basic Data Cleaning / Preprocessing ---
print("\n--- 3. Data Cleaning / Preprocessing ---")

# --- Decide whether to keep Message column ---
# If you need the Message column for later extraction (like in the causal script),
# comment out this dropping section. Otherwise, keep it to drop the column.
# print(f"\nDecision: Keeping '{message_col}' for potential later use (e.g., TypeDescription extraction).")
# Drop Redundant Message Column (Uncomment if you DON'T need it later)
# if message_col in df.columns:
#     print(f"\nDropping potentially redundant column: {message_col}")
#     try:
#         df.drop(columns=[message_col], inplace=True)
#         gc.collect() # Explicitly call garbage collector
#     except KeyError:
#         print(f"Column {message_col} already removed or does not exist.")
# --- End Message Column Handling ---


# Handle Missing Values
print("\nMissing values per column (before handling):")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
# --- Strategy: Drop rows with NA in critical columns ---
# Define critical columns needed for core analysis (ensure timestamp/ingestion are included if needed for datetime)
critical_na_cols = [healthcode_col, type_col, track_section_col, timestamp_col, ingestion_date_col, vehicle_col]
# Ensure these critical columns actually exist before trying to drop NA based on them
critical_na_cols = [col for col in critical_na_cols if col in df.columns]
if critical_na_cols:
    print(f"\nDropping rows with NAs in any of: {critical_na_cols}...")
    start_rows = df.shape[0]
    # Convert potential boolean Series to object before dropping NA if necessary
    # Example: if 'col_name' can be boolean with NA -> df['col_name'] = df['col_name'].astype(object)
    df.dropna(subset=critical_na_cols, inplace=True)
    print(f"Dropped {start_rows - df.shape[0]} rows.")
    if df.empty:
        print("DataFrame is empty after dropping NA rows. Stopping.")
        sys.exit()
else:
    print("Skipping NA drop based on critical columns list.")


# Handle Duplicate Rows
duplicate_rows = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")
if duplicate_rows > 0:
    print("Dropping duplicate rows...")
    df.drop_duplicates(inplace=True)
    print(f"DataFrame shape after dropping duplicates: {df.shape}")
    gc.collect()

# ***** START: DATETIME CONVERSION FIX *****
# --- Inspect Timestamp before conversion ---
print("\n--- Inspecting raw Timestamp data before conversion ---")
raw_timestamps_to_inspect = pd.Series(dtype=object) # Initialize empty series
if timestamp_col in df.columns:
    print(f"First 10 non-null values of '{timestamp_col}':")
    try:
        # Get first 10 non-null unique values for better inspection
        raw_timestamps_to_inspect = df[timestamp_col].dropna().unique()[:10]
        print(raw_timestamps_to_inspect)
    except Exception as e:
        print(f"Could not display head of {timestamp_col}: {e}")
else:
    print(f"Column '{timestamp_col}' not found before conversion attempt.")
# --- End Inspection ---


# --- Convert Data Types (Date/Time) ---
print("\nConverting date/time columns...")
datetime_conversion_successful = False
temp_event_dt_col = "EventDateTime_temp" # Use a temporary column name

# Check if the original timestamp column exists
if timestamp_col in df.columns:
    try:
        # --- Attempt 1: Parse using the specified format ---
        print(f"Attempt 1: Parsing '{timestamp_col}' using explicit format: '{datetime_format}'")
        df[temp_event_dt_col] = pd.to_datetime(df[timestamp_col], format=datetime_format, errors='coerce')

        # Check if Attempt 1 failed for all rows
        if df[temp_event_dt_col].isnull().all():
            print(f"Attempt 1 FAILED for all rows. Explicit format '{datetime_format}' likely incorrect.")
            # --- Attempt 2: Try automatic format inference ---
            print(f"Attempt 2: Parsing '{timestamp_col}' using automatic inference...")
            try:
                # Important: Work on the original timestamp column again
                df[temp_event_dt_col] = pd.to_datetime(df[timestamp_col], errors='coerce', infer_datetime_format=True)
                if df[temp_event_dt_col].isnull().all():
                    print("Attempt 2 FAILED for all rows. Automatic inference could not determine format.")
                    # Give specific advice based on inspection
                    print("\n--- ACTION REQUIRED ---")
                    print(f"Could not parse the '{timestamp_col}' column.")
                    print("Please:")
                    print("1. Examine the 'First 10 non-null values' printed above.")
                    print(f"   Example values seen: {list(raw_timestamps_to_inspect)}")
                    print("2. Determine the correct Python strptime format string for these values.")
                    print("   (e.g., '%Y-%m-%d %H:%M:%S', '%m/%d/%Y %I:%M:%S %p', etc.)")
                    print(f"3. Update the 'datetime_format' variable at the top of the script with the correct format.")
                    print("-----------------------\n")
                    # Optionally exit or continue without the datetime column
                    # sys.exit("Stopping due to datetime parsing failure.")
                    print("Proceeding without 'EventDateTime' column.")

                elif df[temp_event_dt_col].isnull().any():
                     num_failed = df[temp_event_dt_col].isnull().sum()
                     print(f"Attempt 2 SUCCEEDED, but {num_failed} rows still resulted in NaT (likely inconsistent formats).")
                     # Decide how to handle partial failure (e.g., drop NaTs, keep them)
                     datetime_conversion_successful = True # Mark as successful if most parsed
                else:
                     print("Attempt 2 SUCCEEDED using automatic inference.")
                     datetime_conversion_successful = True

            except Exception as e_infer:
                 print(f"Attempt 2 FAILED with an error during inference: {e_infer}")
                 print("Please check the data for severe inconsistencies or errors.")

        elif df[temp_event_dt_col].isnull().any():
            num_failed = df[temp_event_dt_col].isnull().sum()
            print(f"Attempt 1 SUCCEEDED using explicit format, but {num_failed} rows resulted in NaT (check format consistency).")
            # Decide how to handle partial failure
            datetime_conversion_successful = True # Mark as successful if most parsed
        else:
            print(f"Attempt 1 SUCCEEDED using explicit format '{datetime_format}'.")
            datetime_conversion_successful = True

        # --- Final step: If successful, rename temp column and drop originals ---
        if datetime_conversion_successful:
            df.rename(columns={temp_event_dt_col: 'EventDateTime'}, inplace=True)
            print("Created 'EventDateTime' column.")
            # Drop original columns only if EventDateTime was successfully created
            cols_to_drop = []
            if timestamp_col in df.columns: cols_to_drop.append(timestamp_col)
            if ingestion_date_col in df.columns: cols_to_drop.append(ingestion_date_col)
            if cols_to_drop:
                print(f"Dropping original columns: {cols_to_drop}")
                df.drop(columns=cols_to_drop, inplace=True)
        elif temp_event_dt_col in df.columns:
             # Clean up temp column if conversion ultimately failed
             df.drop(columns=[temp_event_dt_col], inplace=True)


    except Exception as e:
        print(f"Unexpected Error during datetime conversion process: {e}")
        # Clean up temp column if it exists after an error
        if temp_event_dt_col in df.columns:
             df.drop(columns=[temp_event_dt_col], inplace=True, errors='ignore')

    finally:
        gc.collect()

else:
     # Handle case where source timestamp column is missing
     print(f"Skipping datetime conversion because source column '{timestamp_col}' was not found.")

# ***** END: DATETIME CONVERSION FIX *****


# Convert Categorical & Optimize Numerics
print("\nConverting types for memory efficiency (Categorical)...")
# Define the list of columns intended to be categorical
categorical_to_convert = [healthcode_col, type_col, track_section_col, vehicle_col]
for col in categorical_to_convert:
    if col in df.columns:
        try:
            # If column is numeric convert to string first, only if appropriate (TrackSection might be ID-like)
            if pd.api.types.is_numeric_dtype(df[col]):
                 print(f"Converting potentially numeric ID column '{col}' to string before categorizing.")
                 df[col] = df[col].astype(str)

            # Now convert to category
            print(f"Converting column '{col}' to category type.")
            df[col] = df[col].astype('category')
        except Exception as e:
            print(f"Error converting column '{col}' to category: {e}. Skipping.")


print("\nConverting types for memory efficiency (Numeric)...")
# Downcast remaining numerical columns if they exist
# Ensure reduce_mem_usage handles potential NaNs when finding min/max
numerical_cols_to_reduce = [col for col in numerical_cols_base if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
if numerical_cols_to_reduce:
     print(f"Optimizing numeric columns: {numerical_cols_to_reduce}")
     df = reduce_mem_usage(df, verbose=False) # Apply to the whole df or just the slice
else:
     print("No numerical columns specified in 'numerical_cols_base' found to optimize.")

# --- 4. Memory Check After Cleaning ---
print("\n--- 4. Memory Usage After Cleaning ---")
df.info(memory_usage='deep')
gc.collect()


# --- Code from Section 5, 6, 7 (Plotting Disabled) and Section 8 (Saving) would go here ---
# --- ... (Assuming the rest of the script remains as in the previous version with plotting disabled) ... ---

# --- 5. Univariate Analysis (Plotting Disabled) ---
print("\n--- 5. Univariate Analysis (Plotting Disabled) ---")
# (Calculations and prints, no plt or sns calls)
# ... (previous code for section 5 without plotting) ...
# Analyze Datetime Column
if datetime_conversion_successful and 'EventDateTime' in df.columns and df['EventDateTime'].notna().any():
    print("\nAnalyzing EventDateTime column (plotting disabled)...")
    try:
        # Resample requires a DatetimeIndex
        event_counts_daily = df.set_index('EventDateTime')['EventDateTime'].resample('D').size()
        print("Daily event counts calculated (plotting disabled).")
        print(event_counts_daily.head())
        del event_counts_daily # Clean up resampled series
    except Exception as e:
        print(f"Could not perform time series resampling/stats: {e}")
    # finally:
        # plt.clf(); plt.close(); gc.collect() # Not needed without plots
elif 'EventDateTime' not in df.columns:
     print("\nSkipping EventDateTime analysis because 'EventDateTime' column was not created.")
elif df['EventDateTime'].isnull().all():
     print("\nSkipping EventDateTime analysis because the column exists but contains only NaT values.")


# --- 6. Identify Top Categories for Bivariate ---
print("\n--- 6. Identifying Top Categories for Bivariate Analysis ---")
# (Code remains the same as it doesn't plot)
# ... (previous code for section 6) ...
try:
    if healthcode_col in df.columns and df[healthcode_col].dtype.name == 'category':
        top_healthcodes = df[healthcode_col].value_counts().nlargest(N_TOP_HEALTHCODES).index
        print(f"- Top {N_TOP_HEALTHCODES} Healthcodes identified.")
    else:
        print(f"Warning: Cannot identify top Healthcodes. Column '{healthcode_col}' missing or not categorical.")

    if type_col in df.columns and df[type_col].dtype.name == 'category':
        top_types = df[type_col].value_counts().nlargest(M_TOP_TYPES).index
        print(f"- Top {M_TOP_TYPES} Types identified.")
    else:
        print(f"Warning: Cannot identify top Types. Column '{type_col}' missing or not categorical.")

    if track_section_col in df.columns and df[track_section_col].dtype.name == 'category':
        top_track_sections = df[track_section_col].value_counts().nlargest(K_TOP_TRACK_SECTIONS).index
        print(f"- Top {K_TOP_TRACK_SECTIONS} Track sections identified.")
        track_section_analysis_possible = True # Set flag to True
    else:
        print(f"Warning: Cannot identify top Track sections. Column '{track_section_col}' missing or not categorical.")
        track_section_analysis_possible = False # Keep flag False

except Exception as e:
    print(f"Error identifying top categories for bivariate analysis: {e}")
    print("Bivariate analysis may be incomplete.")


# --- 7. Bivariate Analysis (Plotting Disabled) ---
print("\n--- 7. Bivariate Analysis (Plotting Disabled) ---")
# (Calculations and prints, no plt or sns calls)
# ... (previous code for section 7 without plotting) ...
# Updated Function to generate and print crosstabs (Plotting Disabled)
def calculate_print_crosstab(df_data, index_col, columns_col, top_index_cats, top_column_cats, N_index, M_columns):
    if top_index_cats is None or top_column_cats is None: return
    if not (index_col in df_data.columns and columns_col in df_data.columns): return
    # Check if columns exist AND are categorical before proceeding
    if not (df_data[index_col].dtype.name == 'category' and df_data[columns_col].dtype.name == 'category'):
        print(f"Skipping crosstab for {index_col} vs {columns_col} as one or both are not categorical type.")
        return

    print(f"\n--- Analyzing Relationship: {index_col} vs {columns_col} ---")
    idx_filter = df_data[index_col].isin(top_index_cats)
    col_filter = df_data[columns_col].isin(top_column_cats)
    df_filtered_local = df_data.loc[idx_filter & col_filter, [index_col, columns_col]].copy() # Use .loc and copy

    if not df_filtered_local.empty:
        try:
            # Ensure categories are set for consistent crosstab output
            df_filtered_local[index_col] = pd.Categorical(df_filtered_local[index_col], categories=top_index_cats, ordered=True)
            df_filtered_local[columns_col] = pd.Categorical(df_filtered_local[columns_col], categories=top_column_cats, ordered=True)

            print(f"Calculating Crosstab (Counts)...")
            crosstab_counts = pd.crosstab(df_filtered_local[index_col], df_filtered_local[columns_col], dropna=False)
            print(crosstab_counts)

            print(f"Calculating Crosstab (% within {columns_col})...")
            crosstab_norm = pd.crosstab(df_filtered_local[index_col], df_filtered_local[columns_col], normalize='columns', dropna=False) * 100
            print(crosstab_norm.round(1))
            del crosstab_counts, crosstab_norm # Clean up
        except Exception as e:
            print(f"Error generating crosstab calculation for {index_col} vs {columns_col}: {e}")
    else:
        print(f"No overlapping data found for the selected top {index_col} and {columns_col}.")
    del df_filtered_local # Clean up local copy
    gc.collect()

# Generate the crosstab calculations conditionally
if top_healthcodes is not None and top_types is not None:
    calculate_print_crosstab(df, healthcode_col, type_col, top_healthcodes, top_types, N_TOP_HEALTHCODES, M_TOP_TYPES)
else:
    print("\nSkipping Healthcode vs Type crosstab calculation.")

if track_section_analysis_possible and top_healthcodes is not None and top_track_sections is not None:
    calculate_print_crosstab(df, healthcode_col, track_section_col, top_healthcodes, top_track_sections, N_TOP_HEALTHCODES, K_TOP_TRACK_SECTIONS)
else:
    print("\nSkipping Healthcode vs Track section crosstab calculation.")

if track_section_analysis_possible and top_types is not None and top_track_sections is not None:
    calculate_print_crosstab(df, type_col, track_section_col, top_types, top_track_sections, M_TOP_TYPES, K_TOP_TRACK_SECTIONS)
else:
    print("\nSkipping Type vs Track section crosstab calculation.")


# --- 8. Save Processed Data ---
print("\n--- 8. Saving Processed Data ---")
# Note: This saves the final 'df' DataFrame after cleaning and processing from this script.
if 'df' in locals() and not df.empty: # Check if df exists and is not empty
    try:
        print(f"Saving processed DataFrame (shape: {df.shape}) to: {output_file_path}")
        df.to_pickle(output_file_path)
        print("DataFrame saved successfully as a pickle file.")
    except Exception as e:
        print(f"Error saving DataFrame to pickle file: {e}")
else:
    print("Skipping save because the DataFrame 'df' is empty or does not exist.")


print("\n--- EDA Script Finished ---")
```

```python
import pandas as pd
import numpy as np
import dowhy
from dowhy import CausalModel
import logging
import gc
from sklearn.preprocessing import StandardScaler # Import StandardScaler

# Optional: Suppress excessive logging from libraries if needed
# logging.getLogger("dowhy").setLevel(logging.WARNING)

# --- 0. Load Processed Data ---
# Assume 'df' is loaded here (the fully processed DataFrame from EDA)
try:
    _ = df.shape # Check if df is defined
    print(f"Using pre-loaded DataFrame with shape: {df.shape}")
    # Ensure EventDateTime is actually datetime type after loading
    if 'EventDateTime' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['EventDateTime']):
        print("Warning: EventDateTime not loaded as datetime. Attempting conversion...")
        df['EventDateTime'] = pd.to_datetime(df['EventDateTime'], errors='coerce')

except NameError:
    print("Error: DataFrame 'df' not found.")
    print("Please load your processed data (output from the full EDA script) first.")
    exit()
except Exception as e:
    print(f"Error accessing DataFrame 'df': {e}")
    exit()

# --- Define Columns Needed ---
healthcode_col = 'Healthcode'
type_col = 'Type'
track_section_col = 'TrackSection' # Use the verified name
vobc_id_col = 'VobcId'
datetime_col = 'EventDateTime'
loop_num_col = 'LoopNum'
# Add other numerical columns if they should be confounders or scaled
numerical_confounders_to_scale = [loop_num_col] # Keep scaling LoopNum for consistency


# --- Handle Missing Timestamps (Ensure this ran in EDA) ---
print("\n--- 0.5 Handling Missing Timestamps ---")
if datetime_col in df.columns:
    initial_rows = df.shape[0]
    nat_count = df[datetime_col].isnull().sum()
    if nat_count > 0:
        percentage_nat = (nat_count / initial_rows) * 100 if initial_rows > 0 else 0
        print(f"Found {nat_count} rows ({percentage_nat:.2f}%) with NaT in '{datetime_col}'.")
        print(f"Dropping rows with NaT in '{datetime_col}'.")
        df.dropna(subset=[datetime_col], inplace=True)
        print(f"DataFrame shape after dropping NaT rows: {df.shape}")
        remaining_nats_df = df[datetime_col].isnull().sum()
        print(f"Verification: NaT count in df['{datetime_col}'] after drop = {remaining_nats_df}")
        if remaining_nats_df > 0: print("ERROR: NaT values still present after dropna!")
        if df.empty: print("Error: DataFrame empty after NaT drop."); exit()
    else:
        print(f"No NaT values found in '{datetime_col}'.")
else:
    print(f"Warning: '{datetime_col}' column not found.")


# --- 1. Define Outcome Variable ---
# *** CHANGE: Outcome is now Healthcode '4F' (Door Failures) ***
target_healthcode = '4F' # Changed to Door Failure code
outcome_var = f'IsHC{target_healthcode}' # e.g., IsHC4F
print(f"\n--- 1. Defining Outcome Variable ({outcome_var} for HC {target_healthcode}) ---")

# Ensure Healthcode column is string or category
df[healthcode_col] = df[healthcode_col].astype(str)
df[outcome_var] = (df[healthcode_col] == target_healthcode).astype(int)
print(f"Created '{outcome_var}' column. Number of HC {target_healthcode} events found: {df[outcome_var].sum()}")

# Check if there are enough outcome events
if df[outcome_var].sum() < 50: # Arbitrary threshold
    print(f"Warning: Very few HC {target_healthcode} events found. Causal estimation might be unstable or fail.")
elif df[outcome_var].sum() == 0:
     print(f"Error: No HC {target_healthcode} events found (after potential NaT drop). Cannot proceed.")
     exit()


# --- 2. Define Treatment Variable and Filter Data ---
# *** CHANGE: Treatment is now TrackSection '50345' vs Baseline '0' ***
print(f"\n--- 2. Defining Treatment Variable (TrackSection {target_healthcode} vs Baseline) ---")
target_track_sections = ['50345'] # << CHANGED: Focus only on 50345
baseline_track_section = '0' # Use common section '0' as baseline/control

# Ensure TrackSection column is string or category
df[track_section_col] = df[track_section_col].astype(str)

# Filter data to include only the sections being compared
sections_to_include = target_track_sections + [baseline_track_section]
df_filtered = df[df[track_section_col].isin(sections_to_include)].copy()

# Check if enough data remains
if df_filtered.shape[0] == 0:
     print(f"Error: No data found for TrackSections {sections_to_include} (after potential NaT drop). Cannot proceed.")
     exit()

# Create the binary treatment column (1 if target section, 0 if baseline section)
temp_treatment_col = 'IsTargetTrackSection' # Temporary name
df_filtered[temp_treatment_col] = df_filtered[track_section_col].isin(target_track_sections).astype(int)

# --- Rename column to 'treatment' and update treatment_var ---
treatment_var = 'treatment' # Use default name for dowhy variable
df_filtered.rename(columns={temp_treatment_col: treatment_var}, inplace=True)
print(f"Renamed treatment column to '{treatment_var}'")
# -----------------------------------------------------------------

print(f"Filtered data shape: {df_filtered.shape}")
treatment_count = df_filtered[treatment_var].sum()
control_count = len(df_filtered) - treatment_count
print(f"Treatment group (TrackSection in {target_track_sections}): {treatment_count} rows")
print(f"Control group (TrackSection={baseline_track_section}): {control_count} rows")

# Check if both groups have sufficient size
min_group_size = 50
if treatment_count < min_group_size or control_count < min_group_size:
    print(f"Warning: Treatment or Control group has fewer than {min_group_size} samples. Results may be unstable.")

# Check outcome occurrence in both groups
outcome_in_treatment = df_filtered[df_filtered[treatment_var] == 1][outcome_var].sum()
outcome_in_control = df_filtered[df_filtered[treatment_var] == 0][outcome_var].sum()
print(f"Number of HC {target_healthcode} events in Treatment Group: {outcome_in_treatment}")
print(f"Number of HC {target_healthcode} events in Control Group: {outcome_in_control}")
if outcome_in_treatment == 0 or outcome_in_control == 0:
     print(f"CRITICAL Warning: The outcome event (HC {target_healthcode}) does not occur in both the treatment and control groups after filtering.")
     print("Causal estimation will likely fail or produce unreliable results due to lack of overlap.")
     # Consider stopping if overlap is required: exit()


# --- 3. Define and Prepare Confounders ---
print("\n--- 3. Defining and Preparing Confounders ---")
# Confounders: Variables affecting BOTH TrackSection choice AND HC 4F risk
# Type, VobcId, time, LoopNum are kept as potential confounders
time_confounders = []
if datetime_col in df_filtered.columns and pd.api.types.is_datetime64_any_dtype(df_filtered[datetime_col]):
    print(f"Creating time features from '{datetime_col}'...")
    nats_before_dt_access = df_filtered[datetime_col].isnull().sum()
    print(f"Verification: NaT count in df_filtered['{datetime_col}'] before .dt access = {nats_before_dt_access}")
    if nats_before_dt_access > 0: print(f"WARNING: NaT values detected.")
    try:
        df_filtered['hour_of_day'] = df_filtered[datetime_col].dt.hour.astype('Int8')
        df_filtered['day_of_week'] = df_filtered[datetime_col].dt.dayofweek.astype('Int8')
        time_confounders = ['hour_of_day', 'day_of_week']
        print("Added time-based confounders (as nullable Int8): hour_of_day, day_of_week")
        hour_nas = df_filtered['hour_of_day'].isnull().sum()
        day_nas = df_filtered['day_of_week'].isnull().sum()
        if hour_nas > 0 or day_nas > 0: print(f"WARNING: Created time features contain <NA> values (hour: {hour_nas}, day: {day_nas}).")
    except Exception as e:
        print(f"Error creating time features: {e}. Skipping them.")
        time_confounders = []
else:
    print(f"Warning: '{datetime_col}' column not found or not datetime type. Skipping time-based confounders.")

base_confounder_names = [type_col, vobc_id_col, loop_num_col] + time_confounders
confounder_names = [c for c in base_confounder_names if c in df_filtered.columns]
print(f"Final confounders list: {confounder_names}")


# --- Scale Numerical Confounders ---
confounders_to_scale_present = [c for c in numerical_confounders_to_scale if c in confounder_names]
if confounders_to_scale_present:
    print(f"Scaling numerical confounders: {confounders_to_scale_present}...")
    # Convert columns potentially containing pd.NA (from Int8) to float before scaling/imputation
    for col in confounders_to_scale_present:
        if pd.api.types.is_integer_dtype(df_filtered[col]) and df_filtered[col].isnull().any():
             print(f"Converting column '{col}' to float due to NaNs before scaling.")
             df_filtered[col] = df_filtered[col].astype(float)
        # Check if any NaNs remain after potential float conversion (shouldn't if loop_num is original int)
        if df_filtered[col].isnull().any():
            print(f"Warning: Column '{col}' contains NaNs before scaling. Imputing with median.")
            median_val = df_filtered[col].median()
            df_filtered[col].fillna(median_val, inplace=True)

    try:
        scaler = StandardScaler()
        df_filtered[confounders_to_scale_present] = scaler.fit_transform(df_filtered[confounders_to_scale_present])
        print("Scaling complete.")
    except Exception as e:
         print(f"Error during scaling: {e}. Check columns: {confounders_to_scale_present}")
else:
    print("No specified numerical confounders found to scale.")


# --- Check for NaNs in columns used by the model ---
print("\nChecking for NaNs in model columns...")
# Include type_col, vobc_id_col if they are confounders
model_cols = [outcome_var, treatment_var] + confounder_names
model_cols = [col for col in model_cols if col in df_filtered.columns]
if not model_cols: print("Error: No columns identified for the model."); exit()

# Convert nullable integers to float before isnull check if needed, but isnull should work
nans_in_model_cols = df_filtered[model_cols].isnull().sum()
print("NaN counts:")
print(nans_in_model_cols[nans_in_model_cols > 0])

if nans_in_model_cols.sum() > 0:
    print("\nError: NaNs detected in columns required for causal model!")
    print(f"Columns with NaNs: {nans_in_model_cols[nans_in_model_cols > 0].index.tolist()}")
    print("Handle these NaNs (e.g., imputation or dropping) before proceeding.")
    # Example: Dropping rows with any NaN in essential model columns
    print("Attempting to drop rows with remaining NaNs in model columns...")
    initial_row_count = df_filtered.shape[0]
    df_filtered.dropna(subset=model_cols, inplace=True)
    print(f"Dropped {initial_row_count - df_filtered.shape[0]} rows.")
    if df_filtered.empty:
         print("Error: DataFrame became empty after dropping remaining NaNs.")
         exit()
    # Verify again
    nans_after_drop = df_filtered[model_cols].isnull().sum().sum()
    if nans_after_drop > 0:
         print("ERROR: NaNs still present after second drop attempt. Exiting.")
         exit()
    else:
         print("NaNs handled by dropping rows.")
else:
    print("No NaNs found in model columns.")


# --- 4. DoWhy Causal Analysis ---

# --- Step 1: Model ---
print("\n--- 4.1. Modeling Causal Assumptions ---")
# Check if essential columns exist before creating model
required_model_cols = {treatment_var, outcome_var}.union(set(confounder_names))
missing_model_cols = required_model_cols - set(df_filtered.columns)
if missing_model_cols:
    print(f"Error: Missing columns needed for model: {missing_model_cols}")
    exit()

# Ensure treatment and outcome are binary
if not df_filtered[treatment_var].isin([0, 1]).all():
    print(f"Error: Treatment variable '{treatment_var}' is not binary (0/1).")
    exit()
if not df_filtered[outcome_var].isin([0, 1]).all():
    print(f"Error: Outcome variable '{outcome_var}' is not binary (0/1).")
    exit()

# Convert categorical confounders to dummy variables for propensity score methods
categorical_confounders = [c for c in confounder_names if df_filtered[c].dtype.name == 'category' or pd.api.types.is_object_dtype(df_filtered[c])]
if categorical_confounders:
    print(f"Converting categorical confounders to dummies: {categorical_confounders}")
    # Keep track of original columns to remove later if desired
    original_cols = df_filtered.columns.tolist()
    df_filtered = pd.get_dummies(df_filtered, columns=categorical_confounders, drop_first=True, dummy_na=False) # drop_first avoids multicollinearity, dummy_na handles potential explicit NaN categories
    # Update confounder list to include new dummy columns and remove original categoricals
    dummy_cols = list(set(df_filtered.columns) - set(original_cols))
    confounder_names = [c for c in confounder_names if c not in categorical_confounders] + dummy_cols
    print(f"Updated confounder list with dummies: {confounder_names[:5]}... (+ {len(confounder_names)-5} more)")


model = CausalModel(
    data=df_filtered,
    treatment=treatment_var,      # 'treatment' (IsTargetTrackSection: 0 or 1)
    outcome=outcome_var,          # IsHC4F (0 or 1)
    common_causes=confounder_names # Updated list possibly including dummies
)
print("Model Summary:")
# Suppress graphviz display issues for now
# model.view_model()
# try:
#     from IPython.display import Image, display
#     display(Image(filename="causal_model.png"))
# except Exception as e:
#     print(f"Could not display graph: {e}.")
print("Model assumptions (Common Causes - first 10):", model._common_causes[:10])


# --- Step 2: Identify ---
print("\n--- 4.2. Identifying Causal Effect ---")
try:
    # --- FIX: Removed unsupported argument ---
    identified_estimand = model.identify_effect()
    # ---------------------------------------
    print("Identified Estimand:")
    print(identified_estimand)
except Exception as e:
     print(f"Error during effect identification: {e}")
     print("Cannot proceed without identified effect.")
     exit()

# --- Function to Run Estimation and Refutation ---
def run_estimation_and_refutation(model, estimand, method_name, method_params={}):
    print(f"\n--- Running Estimator: {method_name} ---")
    estimate = None
    try:
        estimate = model.estimate_effect(
            estimand,
            method_name=method_name,
            target_units="ate",
            method_params=method_params
        )
        print(f"\nCausal Estimate using {method_name}:")
        print(estimate)
        print(f"Interpretation: Estimated ATE = {estimate.value:.4f}. Effect of target TrackSection ({target_track_sections}) vs baseline ({baseline_track_section}) on HC {target_healthcode}.")

        print(f"\n--- Refuting Estimate from {method_name} ---")
        refutation_successful = True
        # Refuter 1
        try:
            print("\nRefuting with Add Unobserved Common Cause...")
            refute_rcc = model.refute_estimate(estimand, estimate, method_name="add_unobserved_common_cause")
            print(refute_rcc)
        except Exception as e_ref: print(f"  Add Unobserved Common Cause failed: {e_ref}"); refutation_successful = False
        # Refuter 2
        try:
            print("\nRefuting with Placebo Treatment...")
            refute_placebo = model.refute_estimate(estimand, estimate, method_name="placebo_treatment_refuter", placebo_type="permute", num_simulations=50) # Reduced sims for speed
            print(refute_placebo)
        except Exception as e_ref: print(f"  Placebo Treatment failed: {e_ref}"); refutation_successful = False
        # Refuter 3
        try:
            print("\nRefuting with Data Subsets...")
            refute_subset = model.refute_estimate(estimand, estimate, method_name="data_subset_refuter", subset_fraction=0.8, num_simulations=50) # Reduced sims for speed
            print(refute_subset)
        except Exception as e_ref: print(f"  Data Subset failed: {e_ref}"); refutation_successful = False

        if refutation_successful: print(f"\nRefutation attempts completed for {method_name}.")
        else: print(f"\nSome refutation attempts failed for {method_name}.")

    except Exception as e_est:
        print(f"\nError during estimation with {method_name}: {e_est}")
        estimate = None # Ensure estimate is None if estimation failed

    return estimate


# --- Step 3 & 4 Combined: Estimate and Refute with Multiple Methods ---
print("\n--- 4.3 & 4.4 Combined: Estimating and Refuting Causal Effect (ATE) ---")

results = {} # Dictionary to store results

# Method 1: Propensity Score Matching
results['matching'] = run_estimation_and_refutation(
    model, identified_estimand, "backdoor.propensity_score_matching"
    # Add specific matching params if needed: , method_params={'caliper': 0.1}
)

# Method 2: Propensity Score Weighting (Inverse Probability of Treatment Weighting - IPTW)
results['weighting'] = run_estimation_and_refutation(
    model, identified_estimand, "backdoor.propensity_score_weighting"
    # Add specific weighting params if needed: , method_params={'weighting_scheme': 'ips_weight'}
)

# Method 3 (Optional): Propensity Score Stratification
# results['stratification'] = run_estimation_and_refutation(
#     model, identified_estimand, "backdoor.propensity_score_stratification",
#     method_params={"num_strata": 30, "clipping_threshold": 4}
# )


# --- 5. Summary and Practical Investigation Prompt ---
print("\n--- 5. Summary and Next Steps ---")
print(f"Causal Question: Effect of TrackSection {target_track_sections} vs {baseline_track_section} on HC {target_healthcode} (Door Failure)")

effect_estimates = []
print("\nSummary of Estimates:")
for method, estimate in results.items():
    if estimate:
        print(f"- {method.capitalize()}: ATE = {estimate.value:.4f}")
        effect_estimates.append(estimate.value)
    else:
        print(f"- {method.capitalize()}: Estimation Failed")

if not effect_estimates:
    print("\nNo successful causal estimates obtained.")
else:
    avg_effect = np.mean(effect_estimates)
    print(f"\nAverage successful estimate: ATE = {avg_effect:.4f}")
    # Assess magnitude (this is subjective)
    if abs(avg_effect) > 0.05: # Example threshold for 'large' effect
         print("\n>> The estimated average effect size appears potentially significant.")
    else:
         print("\n>> The estimated average effect size appears small.")

# --- Practical Investigation Prompt ---
print("\n--- Next Steps: Practical Investigation ---")
print(f"Based on the causal analysis results (check estimates and refutation robustness above),")
print(f"consider a practical investigation into why TrackSection {target_track_sections[0]} might influence Door Failures (HC {target_healthcode}):")
print("\nPotential Areas to Investigate:")
print(f"- Environmental Factors: Are there specific weather patterns (rain, ice, heat affecting expansion/contraction), unusual vibrations, or debris common to TrackSection {target_track_sections[0]}?")
print(f"- Track Conditions: Does this section have unique features like tight curves, steep gradients, specific switch types, rail joint conditions, or ballast issues that could stress door mechanisms?")
print(f"- Equipment Issues: Are certain VOBC IDs (vehicles) or vehicle types more prone to door failures, AND do they disproportionately operate on TrackSection {target_track_sections[0]}? (Check VobcId confounder relationship)")
print(f"- Operational Patterns: Are there specific operational procedures, speeds, stopping/starting frequencies, or dwell times at platforms within or near TrackSection {target_track_sections[0]}?")
print(f"- Maintenance Records: Do maintenance logs for vehicles operating on {target_track_sections[0]} show different patterns or frequencies of door-related repairs compared to baseline section {baseline_track_section}?")
print("- Sensor / Transponder Issues: Could issues with trackside equipment (loops, transponders) in this section indirectly lead to logs or vehicle behaviour triggering door error codes?")
print("\nCollaboration with field engineers, maintenance teams, and operations specialists is recommended.")
# --- End Practical Investigation Prompt ---

print("\n--- Causal Analysis Script Finished ---")
gc.collect()
```

```python
import pandas as pd
import numpy as np
import dowhy
from dowhy import CausalModel
import logging
import gc
from sklearn.preprocessing import StandardScaler # Import StandardScaler

# Optional: Suppress excessive logging from libraries if needed
# logging.getLogger("dowhy").setLevel(logging.WARNING)

# --- 0. Load Processed Data ---
# Assume 'df' is loaded here (the fully processed DataFrame from EDA)
try:
    _ = df.shape # Check if df is defined
    print(f"Using pre-loaded DataFrame with shape: {df.shape}")
    if 'EventDateTime' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['EventDateTime']):
        print("Warning: EventDateTime not loaded as datetime. Attempting conversion...")
        df['EventDateTime'] = pd.to_datetime(df['EventDateTime'], errors='coerce')
except NameError:
    print("Error: DataFrame 'df' not found.")
    print("Please load your processed data (output from the full EDA script) first.")
    exit()
except Exception as e:
    print(f"Error accessing DataFrame 'df': {e}")
    exit()

# --- Define Columns Needed ---
healthcode_col = 'Healthcode'
type_col = 'Type'
track_section_col = 'TrackSection'
vobc_id_col = 'VobcId'
datetime_col = 'EventDateTime'
loop_num_col = 'LoopNum'
numerical_confounders_to_scale = [loop_num_col]


# --- Handle Missing Timestamps ---
print("\n--- 0.5 Handling Missing Timestamps ---")
if datetime_col in df.columns:
    initial_rows = df.shape[0]
    nat_count = df[datetime_col].isnull().sum()
    if nat_count > 0:
        percentage_nat = (nat_count / initial_rows) * 100 if initial_rows > 0 else 0
        print(f"Found {nat_count} rows ({percentage_nat:.2f}%) with NaT in '{datetime_col}'.")
        print(f"Dropping rows with NaT in '{datetime_col}'.")
        df.dropna(subset=[datetime_col], inplace=True)
        print(f"DataFrame shape after dropping NaT rows: {df.shape}")
        remaining_nats_df = df[datetime_col].isnull().sum()
        print(f"Verification: NaT count in df['{datetime_col}'] after drop = {remaining_nats_df}")
        if remaining_nats_df > 0: print("ERROR: NaT values still present after dropna!")
        if df.empty: print("Error: DataFrame empty after NaT drop."); exit()
    else:
        print(f"No NaT values found in '{datetime_col}'.")
else:
    print(f"Warning: '{datetime_col}' column not found. Cannot proceed with season-based analysis.")
    exit()


# --- 1. Define Outcome Variable ---
# *** Outcome is Slip/Slide related Healthcodes ***
slip_slide_codes = ['43', '56', '5C']
outcome_var = 'IsSlipSlide'
print(f"\n--- 1. Defining Outcome Variable ({outcome_var}) ---")

df[healthcode_col] = df[healthcode_col].astype(str)
df[outcome_var] = (df[healthcode_col].isin(slip_slide_codes)).astype(int)
print(f"Created '{outcome_var}' column. Number of Slip/Slide events found: {df[outcome_var].sum()}")

if df[outcome_var].sum() < 50:
    print(f"Warning: Very few {outcome_var} events found. Causal estimation might be unstable or fail.")
elif df[outcome_var].sum() == 0:
     print(f"Error: No {outcome_var} events found. Cannot proceed.")
     exit()


# --- 2. Define Treatment Variable (Seasonality) and Filter Data ---
print("\n--- 2. Defining Treatment Variable (Autumn vs Summer) ---")
# Feature Engineering for Season
df['month'] = df[datetime_col].dt.month

# Define Autumn (treatment) and Summer (control) months for Northern Hemisphere (UK context)
# Autumn: September (9), October (10), November (11) -> Leaf fall season
# Summer: June (6), July (7), August (8) -> Baseline comparison
winter_months = [12, 1, 2]
summer_months = [6, 7, 8]

# Filter data to include only these specified seasons
df_filtered = df[df['month'].isin(winter_months + summer_months)].copy()

if df_filtered.shape[0] == 0:
     print(f"Error: No data found for Autumn or Summer months. Cannot proceed.")
     exit()

# Create the binary treatment column (1 if Autumn, 0 if Summer)
treatment_var = 'treatment' # DoWhy default treatment variable name, represents 'IsAutumn'
df_filtered[treatment_var] = df_filtered['month'].isin(winter_months).astype(int)

print(f"Filtered data shape (Autumn & Summer): {df_filtered.shape}")
print(f"Treatment group (Autumn months): {df_filtered[treatment_var].sum()} rows")
print(f"Control group (Summer months): {len(df_filtered) - df_filtered[treatment_var].sum()} rows")

if df_filtered[treatment_var].sum() < 50 or (len(df_filtered) - df_filtered[treatment_var].sum()) < 50:
     print("Warning: Treatment (Autumn) or Control (Summer) group has fewer than 50 samples. Results may be unstable.")


# --- 3. Define and Prepare Confounders ---
print("\n--- 3. Defining and Preparing Confounders ---")
# Confounders: Variables affecting BOTH season (less likely for season itself, but they affect slip risk)
# AND Slip/Slide risk.
# We control for Type, VobcId, TrackSection, LoopNum, and fine-grained time (hour, day_of_week)
# to isolate the additional effect of the broader "season".
time_confounders = []
if datetime_col in df_filtered.columns: # Already checked this exists
    print(f"Creating fine-grained time features from '{datetime_col}'...")
    try:
        df_filtered['hour_of_day'] = df_filtered[datetime_col].dt.hour.astype('Int8')
        df_filtered['day_of_week'] = df_filtered[datetime_col].dt.dayofweek.astype('Int8')
        time_confounders = ['hour_of_day', 'day_of_week']
        print("Added time-based confounders: hour_of_day, day_of_week")
    except Exception as e:
        print(f"Error creating time features: {e}. Skipping them.")
else: # Should not be reached due to earlier exit
    pass


base_confounder_names = [type_col, vobc_id_col, track_section_col, loop_num_col] + time_confounders
confounder_names = [c for c in base_confounder_names if c in df_filtered.columns]
# Ensure 'month' itself (if it was accidentally added) is not a confounder if it defines treatment
if 'month' in confounder_names:
    confounder_names.remove('month')
print(f"Final confounders list: {confounder_names}")


# --- Scale Numerical Confounders ---
confounders_to_scale_present = [c for c in numerical_confounders_to_scale if c in confounder_names]
if confounders_to_scale_present:
    print(f"Scaling numerical confounders: {confounders_to_scale_present}...")
    try:
        scaler = StandardScaler()
        df_filtered[confounders_to_scale_present] = df_filtered[confounders_to_scale_present].astype(float)
        df_filtered.dropna(subset=confounders_to_scale_present, inplace=True)
        if df_filtered.empty: print("Error: DataFrame empty after dropping NaNs from columns to be scaled."); exit()
        df_filtered[confounders_to_scale_present] = scaler.fit_transform(df_filtered[confounders_to_scale_present])
        print("Scaling complete.")
    except Exception as e:
        print(f"Error during scaling: {e}.")
else:
    print("No specified numerical confounders found to scale.")


# --- Check for NaNs in columns used by the model ---
print("\nChecking for NaNs in model columns...")
model_cols = [outcome_var, treatment_var] + confounder_names
model_cols = [col for col in model_cols if col in df_filtered.columns]
if not model_cols: print("Error: No columns identified for the model."); exit()

nans_in_model_cols = df_filtered[model_cols].isnull().sum()
print("NaN counts:")
print(nans_in_model_cols[nans_in_model_cols > 0])

if nans_in_model_cols.sum() > 0:
    print("\nError: NaNs detected in columns required for causal model!")
    print(f"Columns with NaNs: {nans_in_model_cols[nans_in_model_cols > 0].index.tolist()}")
    print("Handle these NaNs before proceeding.")
    exit()
else:
    print("No NaNs found in model columns.")


# --- 4. DoWhy Causal Analysis ---

# --- Step 1: Model ---
print("\n--- 4.1. Modeling Causal Assumptions ---")
# Treatment is 'IsAutumn' (represented by 'treatment' column)
# Outcome is 'IsSlipSlide'
# Confounders are Type, VobcId, TrackSection, LoopNum, hour_of_day, day_of_week
model = CausalModel(
    data=df_filtered,
    treatment=treatment_var,
    outcome=outcome_var,
    common_causes=confounder_names
)
print("Model Summary:")
model.view_model()
try:
    from IPython.display import Image, display
    display(Image(filename="causal_model.png"))
except Exception as e:
    print(f"Could not display graph: {e}.")
    print("Model assumptions (Common Causes):", model._common_causes)

# --- Step 2: Identify ---
print("\n--- 4.2. Identifying Causal Effect ---")
try:
    identified_estimand = model.identify_effect()
    print("Identified Estimand:")
    print(identified_estimand)
except Exception as e:
     print(f"Error during effect identification: {e}")
     print("Cannot proceed without identified effect.")
     exit()

# --- Step 3: Estimate ---
print("\n--- 4.3. Estimating Causal Effect (ATE) ---")
estimate_method = "backdoor.propensity_score_stratification"
estimate_params = {"num_strata": 30, "clipping_threshold": 4}
try:
    print(f"Attempting estimation using {estimate_method}...")
    estimate = model.estimate_effect(
        identified_estimand,
        method_name=estimate_method,
        target_units="ate",
        method_params=estimate_params
    )
    print(f"Causal Estimate using {estimate_method}:")
    print(estimate)
    print(f"\nInterpretation: Estimated average causal effect of Autumn months (vs Summer months) is an ATE of {estimate.value:.4f} on the probability of a Slip/Slide event.")
    initial_estimate_successful = True
except Exception as e:
    print(f"Error during initial estimation ({estimate_method}): {e}")
    initial_estimate_successful = False
    estimate = None

# --- Step 4: Refute ---
# Keep the refutation logic with fallback as before
print("\n--- 4.4. Refuting the Estimate ---")

if initial_estimate_successful:
    estimation_method_used_for_refutation = estimate_method
    refutation_successful = False
    try:
        print(f"\nAttempting refutation using estimator: {estimation_method_used_for_refutation}...")
        # Refuter 1: Add Random Common Cause
        print("\nRefuting with Add Unobserved Common Cause...")
        refute_random_common_cause = model.refute_estimate(identified_estimand, estimate, method_name="add_unobserved_common_cause")
        print(refute_random_common_cause)
        # Refuter 2: Placebo Treatment
        print("\nRefuting with Placebo Treatment...")
        refute_placebo = model.refute_estimate(identified_estimand, estimate, method_name="placebo_treatment_refuter", placebo_type="permute", num_simulations=100)
        print(refute_placebo)
        # Refuter 3: Data Subset Refuter
        print("\nRefuting with Data Subsets...")
        refute_subset = model.refute_estimate(identified_estimand, estimate, method_name="data_subset_refuter", subset_fraction=0.8, num_simulations=100)
        print(refute_subset)
        print("\nRefutation attempts with initial estimator completed.")
        refutation_successful = True
    except ValueError as ve:
        print(f"\nValueError during refutation with {estimation_method_used_for_refutation}: {ve}")
        if "Method requires strata" in str(ve) or "Found array with 0 sample(s)" in str(ve):
            print("Stratification failed during refutation. Attempting fallback...")
            refutation_successful = False
        else:
            print(f"An unexpected ValueError occurred during refutation: {ve}")
            refutation_successful = False
    except Exception as e:
        print(f"\nAn unexpected error occurred during refutation with {estimation_method_used_for_refutation}: {e}")
        refutation_successful = False

    # --- Fallback if Initial Refutation Failed ---
    if not refutation_successful:
        print("\n--- Attempting Fallback Estimation and Refutation using Matching ---")
        estimate_method_fallback = "backdoor.propensity_score_matching"
        try:
            print(f"\nRe-estimating using {estimate_method_fallback}...")
            estimate_fallback = model.estimate_effect(identified_estimand, method_name=estimate_method_fallback, target_units="ate")
            print(f"Causal Estimate using {estimate_method_fallback}:"); print(estimate_fallback)
            print(f"\nAttempting refutation using estimator: {estimate_method_fallback}...")
            print("\nRefuting with Add Unobserved Common Cause...")
            refute_rcc_fb = model.refute_estimate(identified_estimand, estimate_fallback, method_name="add_unobserved_common_cause"); print(refute_rcc_fb)
            print("\nRefuting with Placebo Treatment...")
            refute_placebo_fb = model.refute_estimate(identified_estimand, estimate_fallback, method_name="placebo_treatment_refuter", placebo_type="permute", num_simulations=100); print(refute_placebo_fb)
            print("\nRefuting with Data Subsets...")
            refute_subset_fb = model.refute_estimate(identified_estimand, estimate_fallback, method_name="data_subset_refuter", subset_fraction=0.8, num_simulations=100); print(refute_subset_fb)
            print("\nRefutation attempts with fallback estimator completed.")
        except Exception as e_fb:
            print(f"\nError during fallback estimation/refutation ({estimate_method_fallback}): {e_fb}")

else:
    print("\nInitial estimation failed. Skipping refutation.")

print("\n--- Causal Analysis Script Finished ---")
gc.collect()
```
